Построить модель регрессии, дерево решений, оптимизировать параметры и сравнить результаты.

```python
from sklearn.preprocessing import Normalizer
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split

import pandas as pd
```

```python
# Загружаем данные
X = pd.read_csv(load_boston()['filename'], skiprows=1)
```

```python
# Выделяем целевую переменную
y = X['MEDV']
del X['MEDV']
```

```python
# Нормализуем данные
X = Normalizer().fit_transform(X)
```

```python
# Делим на тестовую и обучающую выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Стоим модель регрессии
model = LinearRegression()
```

```python
# Строим дерево решений
tree = DecisionTreeRegressor()
```

```python
# Оптимизируем параметры
model.fit(X_train, y_train)
tree.fit(X_train, y_train)
```

```python
# Сравниваем результаты
print('Качество модели:', model.score(X_test, y_test))
print('Качество дерева:', tree.score(X_test, y_test))
```
